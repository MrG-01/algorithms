```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score, silhouette_score

# Load the Iris dataset
iris = load_iris()

X = iris.data

# The actual species labels for comparison
y_true = iris.target

feature_names = iris.feature_names
target_names = iris.target_names
```

### DBSCAN - Density-Based Spatial Clustering of Applications with Noise

* Finds core samples of high density and expands clusters from them.
* Good for data whose clusters have similar density.
* **`eps` (Epsilon)**
    *  The maximum distance between two samples for them to be considered as in the same neighborhood.
    *  This is a crucial parameter and depends on the scale of the data.
* **`min_samples`**: The number of samples in a neighborhood for a point to be considered as a core point.

### Compare DBSCAN over a range of `eps` and `min_samples`values

* The Adjusted Rand Index (ARI)
    * Evaluates the similarity between the clusterings of an algorithm (DBSCAN), and the ground truth or the actual class labels of the data (Iris species).
    * Considers all pairs of data points and counts how many pairs are:
        * In the same cluster in both the algorithm's output and the ground truth.
        * In different clusters in both the algorithm's output and the ground truth.
    * Interpretation of ARI Score
        * **1.0**: Perfect agreement between the clustering and the ground truth labels.
        * **0**: Clustering is essentially random with respect to the ground truth labels.
* Silhouette Score
    * used to evaluate the quality of clusters formed by a clustering algorithm.
    * `s = (b - a) / max(a, b)`
        * `a` is the average distance from that data point to all other data points within the same cluster.
        * `b` is the average distance from that data point to all data points in the nearest other cluster.
    * Overall Silhouette Score for a clustering is the average of the Silhouette Scores for all the data points in the dataset. **+1 (0.7-1.0** is a strong clustering.

```python
eps_options = [0.3, 0.35, 0.4, 0.45, 0.5]
min_samples_options = [2, 3, 4, 5]

for eps in eps_options:
    for min_samples in min_samples_options:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = list(labels).count(-1)
        ari = adjusted_rand_score(y_true, labels)

        # Calculate silhouette score only if there are more than one cluster
        if n_clusters > 1:
            silhouette = silhouette_score(X, labels)
            print(f"eps={eps}, min_samples={min_samples}, Clusters={n_clusters}, Noise={n_noise}, ARI={ari:.2f}, Silhouette={silhouette:.2f}")
        else:
            print(f"eps={eps}, min_samples={min_samples}, Clusters={n_clusters}, Noise={n_noise}, ARI={ari:.2f}, Silhouette=N/A (only one cluster)")
```

### Run DBSCAN with optimal `eps` and `min_samples`

### Load dataset and run DBSCAN clustering

```python
# Choose DBSCAN parameters
eps_value = 0.4
min_samples_value = 3

# Run DBSCAN
dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
dbscan.fit(X)
labels_dbscan = dbscan.labels_
n_clusters_dbscan = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
n_noise_dbscan = list(labels_dbscan).count(-1)
print(f"DBSCAN Estimated number of clusters: {n_clusters_dbscan}")
print(f"DBSCAN Number of noise points: {n_noise_dbscan}")
```

### Apply PCA for visualization

* The Iris dataset has 4 features so is 4D. Use PCA to reduce dimensions to 2D for scatter plot.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

### Scatterplot with DBSCAN labels

```python
plt.figure(figsize=(12, 5))

# Plot DBSCAN clusters
plt.subplot(1, 2, 1)
unique_labels_dbscan = set(labels_dbscan)
colors_dbscan = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels_dbscan))]

for k, col in zip(unique_labels_dbscan, colors_dbscan):
    if k == -1:
        col = 'k'  # Black color for noise

    class_member_mask = (labels_dbscan == k)
    xy = X_pca[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=10)

plt.title(f'DBSCAN Clustering (eps={eps_value}, min_samples={min_samples_value})')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend([f'Cluster {i}' for i in range(n_clusters_dbscan)] + ['Noise'] if -1 in labels_dbscan else [f'Cluster {i}' for i in range(n_clusters_dbscan)])
plt.grid(True)
```

### Scatter plot with True Iris species labels

```python
# Plot True Labels
plt.subplot(1, 2, 2)
colors_true = ['r', 'g', 'b']  # Red, Green, Blue for the three species

for i, target_name in enumerate(target_names):
    subset = X_pca[y_true == i]
    plt.scatter(subset[:, 0], subset[:, 1], color=colors_true[i], label=target_name, s=50)

plt.title('True Iris Species')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
